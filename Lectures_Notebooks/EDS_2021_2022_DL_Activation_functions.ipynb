```python
import numpy as np
import scipy.special
import matplotlib.pyplot as plt

from math import e, sqrt, sin, cos
```

```python

def binaryStep(x):
    ''' 
    It returns '0' is the input is less then 
    zero otherwise it returns one 
    '''
    return np.heaviside(x, 1)

def sigmoid(x):
    '''
    It returns 1/(1+exp(-x)). where the values lies 
    between zero and one 
    '''

    return 1/(1+np.exp(-x))

def tanh(x):
    ''' 
    It returns the value (1-exp(-2x))/(1+exp(-2x)) and the value 
    returned will be lies in between -1 to 1.
    '''
    return np.tanh(x)


def RELU(x):
    ''' 
    It returns zero if the input is less than 
    zero otherwise it returns the given input. 
    '''
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)

    return x1

def d_ReLU(x):
    d_relu = [1 if value>0 else 0 for value in x]
    return d_relu



def softmax(x):
    ''' 
    Compute softmax values for each sets of scores in x. 
    '''
    return np.exp(x) / np.sum(np.exp(x), axis=0)
```

## Plot the activation functions

```python
fig, ax = plt.subplots(1, 2,figsize=(12,8))

x = np.linspace(-10, 10, 500)

ax[0].plot(x, binaryStep(x), label="step")
ax[0].plot(x, sigmoid(x), label="sigmod")
ax[0].plot(x, tanh(x), label="tanh")
ax[0].plot(x, RELU(x), label="ReLU")
#ax[0].plot(x, softmax(x), label="softmax")
ax[0].axis('tight')
ax[0].grid(True)
ax[0].set_title("Activation function")
ax[0].legend()
ax[0].axes.set_ylim([-1.5,1.5])
ax[0].axes.set_xlim([-5,5])
ax[1].plot(x, np.zeros(len(x)), label="step")
ax[1].plot(x, sigmoid(x)*(1-sigmoid(x)), label="sigmoid")
ax[1].plot(x, 1-tanh(x)**2, label="tanh")
ax[1].plot(x, d_ReLU(x), label="ReLU")

ax[1].axes.set_xlim([-5,5])
ax[1].set_title("Derivatives")
ax[1].grid(True)
ax[1].legend()
plt.suptitle('Few activation Functions')

plt.show()
```
